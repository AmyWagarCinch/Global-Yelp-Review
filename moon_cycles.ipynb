```python
import requests
import json
import pandas as pd
from config import api_key
from config import connection_string
from pprint import pprint
import psycopg2
from sqlalchemy import create_engine
```

```python
url = "http://api.worldweatheronline.com/premium/v1/astronomy.ashx?"
q = "Las Vegas"

blackfridays = ["2013-11-22", "2014-11-28", "2015-11-27", "2016-11-25", "2017-11-24", "2018-11-23"]
christmas_eves = ["2013-12-24", "2014-12-24", "2015-12-24", "2016-12-24", "2017-12-24", "2018-12-24"]
mothers_days = ["2013-05-12", "2014-05-11", "2015-05-10", "2016-05-08", "2017-05-14", "2018-05-13"]
uk_mothers_day = ["2013-03-10", "2014-03-30", "2015-03-15", "2016-03-06", "2017-03-26", "2018-03-11"]
four_20s = ["2013-04-20", "2014-04-20", "2015-04-20", "2016-04-20", "2017-04-20", "2018-04-20"]
valentines_day = ["2013-02-14", "2014-02-14", "2015-02-14", "2016-02-14", "2017-02-14", "2018-02-14"]

dates = blackfridays + christmas_eves + mothers_days + uk_mothers_day + four_20s + valentines_day
```

```python
date_list = []
moon_cycle = []
```

```python
for date in dates:
    
    query_url = url + "q=" + q + "&date=" + str(date) + "&format=json" + "&key=" + api_key
    moon_data = requests.get(query_url).json()      
    
    date_list.append(date)
    moon_cycle.append(moon_data['data']['time_zone'][0]['moon_phase'])
```

```python
moon_dict = {
    "Date" : date_list,
    "Moon Phase": moon_cycle
}
```

```python
moon_df = pd.DataFrame(moon_dict)
```

```python
#use local connection_string from config.py
engine = create_engine(f'postgresql://{connection_string}')
```

```python
moon_df.to_sql(name='moon_cycles', con=engine, if_exists='append', index=False)
```
